```python
import sqlite3
import pandas as pd
```

```python
# Create your connection.
con = sqlite3.connect('database.sqlite')
```

```python
#2.1 Import table Player into pandas DataFrame 
df = pd.read_sql_query("SELECT * FROM Player", con)
```

```python
#2.2 Print first 10 records 
print(df.head(10))
```

```python
#2.3 Print last 15 records 
print(df.tail(15))
```

```python
#2.4 Sort by height (ACS) AND weight (DESC)
df.head(10).sort_values(by=['height', 'weight'], ascending=[True, False])
```

```python
#2.5 Print columns id, player_fifa_api_id and player_name columns
df[['id', 'player_fifa_api_id', 'player_name']].head(10)
```

```python
#2.6 Select records where height is in range [170, 190] and id is in range [500, 700]
df[ (df['height']>=170) & (df['height']<=190) & (df['id']>=500) & (df['id']<=700)]
```

```python
#2.7 Print quantity of players by year of birth

df['birthday'] = df['birthday'].astype('datetime64[ns]')
df['Year'] = df['birthday'].apply(lambda x: "%d" % (x.year))

df.groupby(['Year']).size().reset_index(name='counts')
```

```python
#2.8 Print min height and max weight of players grouped by month of birth 
df['birthday'] = df['birthday'].astype('datetime64[ns]')
df['Month'] = df['birthday'].apply(lambda x: "%d" % (x.month))


df.groupby(['Month']).agg({'height':'min', 'weight':'max'})
```

```python
#2.9 Print top-5 player's first names (without last names)
df['First_name'] = df['player_name'].str.split().str[0]


df1 = df.groupby(['First_name']).size().reset_index(name='counts')

df1.sort_values(by=['counts'], ascending=[False]).head(5)
```

```python
#2.10 Write code using pandas and dataframe you created that will give the same result as the following SQL query:
# 		' SELECT * 
# 			FROM Player 
# 			WHERE weight = 137
# 			UNION ALL
# 			SELECT * 
# 			FROM Player
# 			WHERE height = 170.18'

df1 =  df[df['weight'] == 137]
df2 = df[df['height'] == 170.18]

df_union_all= pd.concat([df1, df2])

print(df_union_all)
```

```python
#2.11 Create 2 new Dataframes by importing tables Country and League 
df_country = pd.read_sql_query("SELECT * FROM Country", con)
df_league = pd.read_sql_query("SELECT * FROM League", con)
```

```python
#Close connection with db

con.close()
```

```python
#2.12 Join dataframes for Country and League  using pandas to get the result of the following SQL query:
# 		' SELECT * 
# 		FROM League l
# 		JOIN Country c on c.id = l.country_id '


df_league.merge(df_country, how='inner', left_on='country_id', right_on = 'id')
```
